## 1)Looping Through an Entire List

```python
magicians = ['alice', 'david', 'carolina'] 
for magician in magicians: 
    print("Hello! My self "+magician +" Magician! Worlds best magician")
```

### b)Using range() to Make a List of Numbers


```python
numbers = list(range(1,6))#for making list of numbers
print(numbers)
```

```python
even_numbers = list(range(2,100,2)) #for printing even numbers
print(even_numbers)
len(even_numbers)
```

```python
squares=[]
for value in range(1,12):
  square=value**2
  squares.append(square)
print(squares)
#1st shortcut method
squares = []
for value in range(1,12):
  squares.append(value**2)
print(squares)
#2nd Shortcut
squares = [value**3 for value in range(1,12)]
print(squares)

#for cube
values = []
for number in range(1,13):
  for value in range(1,12):
    values.append(value**number)
print(values)
```

## Tuples
sometimes you’ll want to create a list of items that cannot 
change. Tuples allow you to do just that. Python refers to values that cannot change as immutable, and an immutable list is called a tuple

```python
diamension_rect=(50,70)
print(diamension_rect[0])
print(diamension_rect[1])
#Looping Through All Values in a Tuple
print("here are the diamension of rectangel")
for d in diamension_rect:
  print(d)
```

```python
#Writing over a Tuple 
#Although you can’t modify a tuple, you can assign a new value to a variable that holds a tuple. So if we wanted to change our dimensions, we could redefine the entire tuple
dimensions = (200, 50)
print("Original dimensions:")
for dimension in dimensions:
 print(dimension)
 
dimensions = (400, 100)
print("\nModified dimensions:")
for dimension in dimensions:
 print(dimension)
```
